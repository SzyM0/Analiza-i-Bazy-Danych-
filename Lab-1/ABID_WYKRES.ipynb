Wykres funkcji $f(x) = x^2 + 5$ w przedziałach $x \in (-1, 1)$, $x \in (-6, 6)$, $x \in (0, 5)$

```python
import matplotlib
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
def fun(x):
    return x**2 + 5


arg1 = np.linspace(-1, 1, 100)
arg2 = np.linspace(-6, 6, 600)
arg3 = np.linspace(0, 5, 500)


fig, ax = plt.subplots(3, 1)
fig.set_size_inches((15,15))

ax[0].plot(arg1, fun(arg1))
ax[0].set_title("(-1 , 1)")
ax[0].set_ylabel("y")
ax[0].set_xlabel("x")
ax[0].grid()

ax[1].plot(arg2, fun(arg2))
ax[1].set_title("(-6, 6)")
ax[1].set_ylabel("y")
ax[1].set_xlabel("x")
ax[1].grid()

ax[2].plot(arg3, fun(arg3))
ax[2].set_title("(0, 5)")
ax[2].set_ylabel("y")
ax[2].set_xlabel("x")
ax[2].grid()

plt.show()
```
